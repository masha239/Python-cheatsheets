```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

**Создание датасета**

```python
# Создадим небольшой датасет с числовыми и категориальными признаками, заведем ему игрушечный индекс

n = 100

cat1 = np.random.randint(1, 3, n)
num4 = np.random.random(n)
num4 = [num4[i] + 2 if cat1[i] == 1 else num4[i] for i in range(n)]

letters = ['a', 'b', 'c']
cat2 = [random.choice(letters) for _ in range(n)]
df = pd.DataFrame({'num1': np.random.uniform(low=0, high=1, size=n),
                   'num2': np.random.normal(loc=5, scale=4, size=n),
                   'num3': range(n),
                   'num4': num4,
                   'cat1': cat1,
                   'cat2': cat2},
                   index=range(0, 2 * n, 2))

df.head(5)
```

**Сразу же научимся его сохранять и загружать**

```python
df.to_csv('df.csv',
          sep=',',  # Разделитель
          na_rep='',  # Как представлять пропущенные значения (str)
          columns=df.columns,  # Можно и поменять имено колонок
          header=True,  # Писать ли имена
          index=False,  # Не сохранять индекс отдельною колонкой. Можно и сохранить
          decimal='.')  # Вдруг понадобится для excel

df1 = pd.read_csv('df.csv')
df1.head()
```

**Заметим, что индекс изменился! Ведь прежний мы не сохранили. Попробуем еще раз**

**Но перед тем, как идти дальше, заметим, что один категориальный (по задумке) признак у нас какой-то очень числовой. Исправим это, заодно обучившись методам apply и map (и банально List comprehension)**

```python
# map

df['cat1'] = df['cat1'].map({1: 'first', 2: 'second'})
df['cat1']
```

```python
# apply

df['cat1'] = df['cat1'].apply(lambda x: x.upper())
df.head()
```

```python
# еще про apply

def my_func(row):
    if row.num3 % 2 == 0:
        return 100
    if row.cat1 == 'FIRST':
        return 200
    return row.num1

example = df.apply(my_func, axis='columns')
example
```

```python
# и третий вариант, конечно

df['cat1_strange'] = [random.choice(x) for x in df['cat1']]
df.head()
```
